# Optional workflow: external aSoCC and external LCA input staging


Import external allocated shares of carrying capacities (aSoCC) and LCA inputs. The function writes
runnable CSV examples and README guides under `project_name` so external files can be staged before
downstream aSoCC, aCC, or ASR calls. The packaged numeric values are dummy demonstration values only
and must be replaced by project specific external values before interpreting results.

Run `set_workspace(...)` before any later function call. It sets the active workspace root
for the current Python session, creates or reuses the workspace, imports package
prerequisites under `data_raw/`, and records setup guidance in `data_raw/summary.log`.

Methodological notes available in

* `data_raw/methodological_notes/` contains notes for definition
  of functional units and allocation methods, prospective allocation, and
  uncertainty sources.
* `data_raw/carrying_capacities/` contains the note for definition of steady
  state and dynamic carrying capacities.


## What the import prepares

External aSoCC inputs are staged under `B1_asocc/external_asocc/`.
Use them when aSoCC, aCC, or ASR should include user provided allocated shares
together with selected pyaesa computed allocation methods. Downstream functions
select those external methods with `external_method`.

External LCA inputs are staged under `A_lca/external_lca/`. Use them when ASR
should use a user provided LCA numerator instead of IO-LCA.
Downstream ASR functions select that route with `lca_args`.

Both input families support deterministic files and Monte Carlo files. The
external aSoCC and external LCA README guides imported by this function explain
retrospective years, prospective SSP scenario files, accepted filenames, and
required columns.


## Import external input folders and runnable examples

The call below writes these folders for the selected `project_name`:

* `B1_asocc/external_asocc/deterministic/` for real deterministic external aSoCC files, including runnable dummy examples.
* `B1_asocc/external_asocc/monte_carlo/` for real Monte Carlo external aSoCC files, including normal CSV and compact CSV dummy examples.
* `B1_asocc/external_asocc/templates/` for the external aSoCC README guide,
  `README_external_asocc_templates.txt`.
* `A_lca/external_lca/deterministic/` for real deterministic external LCA files, including runnable dummy examples.
* `A_lca/external_lca/monte_carlo/` for real Monte Carlo external LCA files, including normal CSV and compact CSV dummy examples.
* `A_lca/external_lca/templates/` for the external LCA README guide,
  `README_external_lca_templates.txt`.

The packaged examples can be run directly with the identifiers documented in the README guides, but
their numeric values are dummy demonstration values. For project specific data, follow the README
filename syntax and save completed files in `deterministic/` or `monte_carlo/` before calling
downstream aSoCC, aCC, or ASR functions.


The function writes external aSoCC and external LCA folders, runnable CSV examples, and
README guides under the selected project. Imported guides and examples are listed in
the function output, and existing staged files are preserved.

## Downstream selectors after files are staged

For external aSoCC, pass the same external method declaration to the downstream
function and to inter method weight functions when custom inter method weights
are used. Example selector forms are:

```python
external_method = {"one_step_methods": ["CO(S)"]}
external_method = {"l1_methods": ["CO-HR(S,cum)"]}
external_method = {"l1_l2_pairs": ["AR(E)::UT(S)"]}
```

For external LCA in ASR, pass the staged version name through `lca_args`; the packaged
example version name is `template` and the Phase C example uses `lcia_method="ef_3.1"`:

```python
lca_args = {
    "external_lca": {"active": True, "version_name": "template"},
}
```

The Phase C ASR notebook includes an example that uses both external aSoCC and
external LCA in the same ASR call.
